```python
def check(string):
    valid = True
    try:
        for c in string:
            if (ord(c) < ord("A") or ord(c) > ord("Z")) and (ord(c) != 32):
                valid = False
    except:
        valid = False
    return valid
```

```python
import pandas as pd
```

```python
df = pd.read_csv("NYC_Dog_Licensing_Dataset.csv", encoding = "latin_1")
```

```python
df.columns
```

```python
df["keep"] = df.apply(lambda row: check(row["AnimalName"]), axis=1)
```

```python
df["keep"].value_counts()
```

```python
df['AnimalName'].value_counts()
```

```python
# Getting rid of unknown and not provided animal names
df = df.loc[((df['AnimalName'] != 'UNKNOWN') & (df['AnimalName'] != 'NAME NOT PROVIDED')), :]
df['AnimalName'].value_counts()
```

```python
# Getting rid of the invalid time in birth
df['AnimalBirthMonth'] = df['AnimalBirthMonth'].astype("str")
df['AnimalBirthMonth'] = df['AnimalBirthMonth'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
# Getting rid of the keep column
df = df.loc[(df['keep'] == True), :]
df = df.drop(columns = 'keep')
df.head()
```

```python
# drop na's in anamal name, gender, birth month, breedname, and borough
df = df.dropna(subset = ['AnimalName', 'AnimalGender', 'AnimalBirthMonth', 'BreedName', 'Borough'])
len(df)
```

```python
#export
df.to_csv('nyc_dog_cleaned.csv', index = False)
```
